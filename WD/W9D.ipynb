An index from your dataframe:

A column from your dataframe:

A row from your dataframe
there are many ways, one example is below
# Drop rows where 'State/Territory' is 'California' or 'Colorado' by filtering with a query
df = df.query("`State/Territory` != 'California' & `State/Territory` != 'Colorado'")

2. How do you iterate over a pandas dataframe?

  print(index, row['State/Territory'])

4. What is data aggregation?  Give an example in Python.
# calculates sum of sales per country
df.groupby('country').agg({'sales': 'sum'})

5. What is GroupBy in Pandas (groupby()). Give an example in Python.
Groupby split the data into groups based on some criteria, and then apply a function to each group independently.
#calculates the mean of the 'sales' column for each country,
df.groupby('country)['sales'].mean()
